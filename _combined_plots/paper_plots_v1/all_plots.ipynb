Notes about plotting best practices:
* Plots should get made and debugged in their own Jupyter notebooks, which then get included here with the %run magic command.
* Constants, plotting defaults, and the list of files to load get set in the do_plot_setup.ipynb file.
* For other files, e.g. time series components, get the path to the base directory with 'plot_default_properties['base_directory']'. This will make sure the file paths are still valid whether it's done on Max's or Richard's computer.
* Try to save figures with descriptive file names like "co_fed_gas_results.png", not "figure_1.png". That way, we don't have to tediously rename a ton of things in code if we insert a figure in the middle; we just have to fix the headers in this file and the references in the paper.
* Let's try to rely more on Github version history to get old plots if we want them, rather than having a big python folder of deprecated plotting functions that are broken or useless in various ways.
